```python
import random
from transformers import pipeline

fill_mask = pipeline("fill-mask", model='bert-base-uncased')
```

```python
def mask_words(essay, mask_percentage):
    words = essay.split()
    num_words = len(words)
    num_to_mask = max(1, int(num_words * mask_percentage))
    masked_indices = random.sample(range(num_words), num_to_mask)
    masked_essay = ' '.join('[MASK]' if i in masked_indices else word for i, word in enumerate(words))
    masked_words = [word for i, word in enumerate(words) if i in masked_indices]
    return masked_essay, masked_words
```

```python
input_essay = "The question of whether university students should have the freedom to study any subject of their choice or if they should be limited to studying subjects deemed useful for the future, such as science and technology, is a topic of debate. Supporters of allowing students to study whatever they like argue that education should be a pursuit of knowledge and personal fulfillment. This approach encourages intellectual curiosity, creativity, and a well-rounded education. It allows students to develop critical thinking skills and a broad understanding of various subjects. On the other hand, proponents of focusing on practical, job-oriented subjects argue that universities should prioritize subjects related to science, technology, engineering, and mathematics (STEM) or other fields with high job demand. This approach ensures that students acquire skills that are directly applicable in the job market, increasing their employability and contributing to economic growth. In my opinion, a balance should be struck between allowing students to pursue their interests and ensuring they acquire practical skills. Students should have the freedom to choose subjects they are passionate about, as this fosters motivation and engagement. However, universities should also offer programs and resources that equip students with the skills and knowledge relevant to the job market. Furthermore, the dynamic nature of the job market should be considered. While science and technology fields currently hold significant demand, the landscape can change rapidly. Therefore, students should be encouraged to develop transferable skills such as critical thinking, communication, problem-solving, and adaptability, regardless of their chosen fields of study. In conclusion, while allowing students the freedom to study subjects of their choice promotes personal fulfillment and a well-rounded education, practical considerations should also be taken into account. A balanced approach that combines academic exploration with practical skill development is essential to prepare students for the future job market. By providing diverse options and emphasizing transferable skills, universities can better equip students to thrive in an ever-changing professional landscape."
```

```python
num_iterations = 7
alignment_scores = []

for i in range(num_iterations):
    # Select percentage of words to mask
    masked_essay, masked_words = mask_words(input_essay, 0.25)

    results = fill_mask(masked_essay)

    top_words = [result[0]['token_str'] for result in results]

    # Calculate the alignment score for each iteration
    matching_count = sum(1 for str1, str2 in zip(masked_words, top_words) if str1.lower() == str2.lower())
    matching_percentage = (matching_count / len(masked_words)) * 100

    alignment_scores.append(matching_percentage)

    # Print the masked words and highest-scoring words for each iteration
    print("Iteration", i + 1)
    print("Masked words:", masked_words)
    print("Highest-scoring words:", top_words)
    print("Alignment score:", matching_percentage)
    print()


average_score = sum(alignment_scores) / num_iterations

# Print the original sentence and the average alignment score
print("Original sentence:", input_essay)
print("Average alignment score:", average_score)
```
